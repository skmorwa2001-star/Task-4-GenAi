# Assignment 4: File Handling

## Task 1: Write sales records to a file

```python
# Sales amounts
sales = [1200, 450, 980, 1500, 3000]

# open file in read mode and write sale
with open('sales_data.txt','w') as file:
    
    for sale in sales:
        file.write(f"{sale}\n")

with open('sales_data.txt','r') as file:
    content=file.read()
    print(content)        
```

## Task 7: Mini Project - Export Discount Prices

```python
# Create a dictionary
prices={
    "Mouse":500,
    "Keyword":800,
    "Monitor":7000,
    "Pendrive":400,
    "Camera":5000
}

# Ask user for discount percentage
discount_percent=float(input("Enter the discount percentage : "))

with open('discount_report.txt','w') as file:
    file.write("Product | Price | Discounted Price\n")

    # calculation for discount
    total_discount=0
    for product in prices:
        original=prices[product]
        discount = original - (original * discount_percent/100)

        total_discount = total_discount + discount

        file.write(product + "|" + str(original)+"|" + str(discount)+"\n")


with open('discount_report.txt','r') as file:
    content=file.read()
    print(content)        


```
